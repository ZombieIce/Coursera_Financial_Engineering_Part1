## Q1 compute the price of zero-coupon bond(ZCB)


1.construct interest binomial lattice

```python
def interest_lattice(r,u,d,q,T):
    lattice = pd.DataFrame()
    for i in range(T+1):
        interest_period = np.zeros(T+1)
        for j in range(i+1):
            interest_period[j] = r * d**j * u**(i-j)
        lattice[i] = interest_period
    return lattice
```

```python
ten_int_lattice = interest_lattice(r0,U,D,Q,10)
ten_int_lattice
```

2.backwardation zero-coupon bond

```python
def back_zcb(i_lattice, face_value):
    n = len(i_lattice.columns)
    zcb_lattice = i_lattice.copy()
    zcb_lattice.iloc[:n,n-1] = face_value
    for i in reversed(range(n-1)):
        for j in range(i+1):
            zcb_lattice.iloc[j,i] = (zcb_lattice.iloc[j,i+1] + zcb_lattice.iloc[j+1,i+1])* 0.5 / (1+i_lattice.iloc[j,i])
    return zcb_lattice
```

```python
zcb_value = back_zcb(ten_int_lattice, 100)
zcb_value
```

```python
def futures_backwardation(value_lattice, T):
    futures_lattice = pd.DataFrame()
    futures_lattice = value_lattice.iloc[:, range(T+1)].copy()
    for i in reversed(range(T)):
        for j in range(i+1):
            futures_lattice.iloc[j,i] = (futures_lattice.iloc[j,i+1] + futures_lattice.iloc[j+1,i+1]) / 2
    return futures_lattice
```

## Q4 Compute the price of an American call option on the same ZCB of the previous three questions. The option has expiration t = 6 and strike = 80.

1.caculate the original call price at each node

```python
def original_call(value_lattice, T , strike):
    option_price = value_lattice.iloc[:,range(T+1)].copy()
    for index in option_price.index:
        for col in option_price.columns:
            option_price.iloc[index, col] = (option_price.iloc[index, col]-strike) if (option_price.iloc[index, col]-strike) > 0 else 0
    return option_price
```

```python
origin_call = original_call(zcb_value, 6, 80)
origin_call
```

2.backwardation call option

```python
def int_discount_backwardation(price_lattice, int_lattice):
    n_col = len(price_lattice.columns)
    interest_lattice = int_lattice.iloc[:,range(n_col)].copy()
    back_lattice = price_lattice.copy()
    for i in reversed(range(n_col - 1)):
        for j in range(i+1):
            price_lattice.iloc[j,i] = ( price_lattice.iloc[j,i+1]+ price_lattice.iloc[j+1,i+1]) * 0.5 / ( 1+ interest_lattice.iloc[j,i])
    return price_lattice
```

```python
call_backwardation = int_discount_backwardation(origin_call, ten_int_lattice)
call_backwardation
```

##  Q5. Compute the initial value of a forward-starting swap that begins at t=1, with maturity t = 10 and a fixed rate of 4.5%. (The first payment then takes place at t = 2 and the final payment takes place at t = 11 as we are assuming, as usual, that payments take place in arrears.) You should assume a swap notional of 1 million and assume that you receive floating and pay fixed.)

1.using 10-period interest lattice

```python
fixed_rate = 0.045
swap_lattice = ten_int_lattice.copy()
swap_lattice
```

2.caculate t=10 value and replace


```python
swap_lattice.iloc[:,10] = (swap_lattice.iloc[:,10] - fixed_rate) / (1 + swap_lattice.iloc[:,10])
swap_lattice
```

3.backwardation swap value

```python
def backwardation_swap(swap):
    n = len(swap.columns)
    new_swap = swap.copy()
    for i in reversed(range(n-1)):
        for j in range(i+1):
            new_swap.iloc[j,i] = ((new_swap.iloc[j,i] - fixed_rate) + (new_swap.iloc[j,i+1] + new_swap.iloc[j+1, i+1]) * 0.5) / (1 + new_swap.iloc[j,i])
    return new_swap
```

```python
swap_price = backwardation_swap(swap_lattice)
swap_price
```
