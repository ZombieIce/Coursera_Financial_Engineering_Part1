```python
import numpy as np
import pandas as pd
```

## Q1

```python
Mark_spot_rate = np.array([3,3.1,3.2,3.3,3.4,3.5,3.55,3.6,3.65,3.7])
A = np.ones_like(Mark_spot_rate) * 3
B = 0.05
P = 0.5
Q = 1 - 0.5
```

```python
def generate_short_rate(a, b, T):
    short_lattice = pd.DataFrame(np.zeros([T,T]))
    for i in range(T):
        for j in range(i+1):
            short_lattice.iloc[j,i] = a[j] * np.exp(b * j)
    return short_lattice
```

```python
short_rate_lattice = generate_short_rate(A,B,10)
short_rate_lattice
```

```python
def generate_element_price(rate_lattice, p, q, T):
    element_price_lattice = pd.DataFrame(np.zeros([T+1, T+1]))
    element_price_lattice.iloc[0,0] = 1
    for i in range(1,T+1):
        for j in range(i+1):
            if j == 0:
                element_price_lattice.iloc[j,i] = q * element_price_lattice.iloc[j,i-1] / (1 + rate_lattice.iloc[j,i-1] / 100)
            if j == i:
                element_price_lattice.iloc[j,i] = p * element_price_lattice.iloc[j-1,i-1] / (1 + rate_lattice.iloc[j-1,i-1] / 100)
            else:
                element_price_lattice.iloc[j,i] = (q * element_price_lattice.iloc[j,i-1] / (1 + rate_lattice.iloc[j,i-1] / 100)) + (p * element_price_lattice.iloc[j-1,i-1] / (1 + rate_lattice.iloc[j-1,i-1] / 100))
    
    return element_price_lattice
```

```python
element_price = generate_element_price(short_rate_lattice,P,Q,10)
element_price
```
