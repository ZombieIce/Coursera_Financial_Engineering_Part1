```python
import numpy as np
import pandas as pd
from scipy.stats import binom
```

```python
r0 = 0.05
U = 1.1
D = 0.9
Q = 0.5
```

1.construct interest binomial lattice

```python
def interest_lattice(r,u,d,q,T):
    lattice = pd.DataFrame()
    for i in range(T+1):
        interest_period = np.zeros(T+1)
        for j in range(i+1):
            interest_period[j] = r * d**j * u**(i-j)
        lattice[i] = interest_period
    return lattice
```

```python
interest_lattice(r0,U,D,Q,10)
```
